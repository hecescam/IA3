Para evaluar el modelo utilizaremos la mtrica conocida como entropa. La entropa de una funcin es mnima en dos puntos diferentes:

- Primero cuando la probabilidad de pertenecer a una determinada clase es 0, es decir no existen muestras de esa clase en el conjunto de datos.

- Cuando la probabilidad de pertenecer a una determinada clase es 1, es decir, todas las muestras de esa clase en el conjunto de datos son iguales.

De esta manera la entropa de una funcin medir la pureza del split, o dicho de otra manera, la capacidad de esta divisin por separar las clases.

```
import seaborn as sns
import pandas as pd
import plotly.express as px
from sklearn.metrics import accuracy_score
from sklearn.metrics import recall_score
from math import log
```

```
dataset = sns.load_dataset("iris")
```

```
dataset.head(10)
```

```
fig = px.scatter(dataset, x="petal_length", y="petal_width" , color="species")
fig.show()
```

```
def transform_classes(y, class_name):
  new_classes = []
  for sample in y:
    if sample == class_name:
      new_sample = class_name
    else:
      new_sample = "Other"
    new_classes.append(new_sample)
  return new_classes

def basic_splitter(df, threshold, class_name, split_type = "lower"):
  Y = transform_classes(df["species"], class_name)
  df["species"] = Y
  if split_type == "lower":
    bool_array = df["petal_length"] < threshold
  elif split_type == "greater":
    bool_array = df["petal_length"] > threshold
  dataset_filtered = df[bool_array]
  return dataset_filtered
```

```
# Ejemplo

threshold =2

df_filtered = basic_splitter(dataset, threshold, 
                             "setosa", split_type = "lower")
```

```
df_filtered.groupby("species").count()
```

```
def entropy_calculation(selected_class, other):
  total = selected_class+other
  prob_class = selected_class/total
  prob_other = other/total
  entropy_class = - prob_class * log(prob_class+0.00001,2)
  entropy_other = - prob_other * log(prob_other+0.00001,2)
  entropy = entropy_class + entropy_other
  return entropy
```

```
entropy_calculation(50,0)
```

## Train/Test Split

Comencemos con uno de los pasos ms importantes en el entrenamiento la seleccin de sets de training y testing. Cuando entrenamos un algoritmo seleccionamos un dataset que contenga la mayor cantidad de informacin posible. Para nuestro algoritmo, este dataset es el universo total de posibilidades que va a tratar de aprender y representa todas las opciones que los datos podran describir. Sin embargo, la realidad es que nuestro dataset nunca contendr todo el universo de posibilidades, por eso nuestro objetivo ser ensear al algoritmo a **generalizar**

Permitidme explicarme:

<center>

</center>

La elipse anterior representa todo el conjunto de datos del que disponemos. Si permitimos al algoritmo aprender este espacio, es posible que encuentre la forma de predecir cada punto en l y obtenga un resultado perfecto en su evaluacin. Sin embargo, este comportamiento es rara vez deseado, puesto que una vez el algoritmo se enfrente a una situacin real, su aprendizaje no le permitir generalizar para la informacin que provenga del exterior de esa elipsis. Resumiendo, nuestro algoritmo no podr generalizar ms all de lo que ha visto hasta el momento. Por eso tradicionalmente se ha dividido el dataset en dos porciones diferentes una para entrenamiento y otra para validacin (training y testing).

La idea es extraer pequeos trozos del dataset de entrenamiento que representen de forma general el espacio de caractersticas original. Nuestro algoritmo seguir siendo entrenando sobre el conjunto ms amplio (Train), pero la evaluacin se realizar sobre el conjunto reducido (Test). Como el algoritmo no ha visto en ningn momento la informacin de este segundo conjunto, las predicciones que realice podrn aportarnos informacin sobre su rendimiento en un escenario fuera de su entrenamiento.

Los hiperparmetros de un algoritmo son un conjunto de valores que controlan el comportamiento del algoritmo. Permiten controlar cualidades como la complejidad que un algoritmo puede adoptar a la hora de aprender el espacio de un conjunto de entrenamiento.

Para nuestro ejemplo actual utilizaremos dos hiperparmetros:

- max_depth: Controla la profundidad mxima del rbol de decisin. Si este parmetro no se define el rbol crecer de forma indefinida hasta alcanzar el mximo nivel de pureza en cada split o hasta que cada split contenga un nmero de muestras igual a min_samples.

- min_samples: Nmero de muestras mnimo permitido para realizar la divisin de un dataset. Si al realizar la divisin el nmero de muestras es inferior a este parmetro el rbol descartara el split.

Estos hiperparmetros se utilizan en este algoritmo para evitar el sobreajuste (overfitting).

El overfitting se produce en un algoritmo cuando un algoritmo es capaz de realizar predicciones muy precisas sobre el set de entrenamiento, pero sin embargo su rendimiento sobre el dataset de validacin es sustancialmente menor. Como habris podido comprobar en el caso anterior, el score en vuestro dataset de test es, en efecto, inferior al dataset de entrenamiento.
